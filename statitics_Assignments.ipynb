```python
import pandas as pd
from scipy.stats import ttest_rel
from scipy.stats import chi2_contingency

data = {
    "Month": ["Month-1", "Month-2", "Month-3", "Month-4", "Month-5", "Month-6", "Month-7", "Month-8", "Month-9"],
    "Region": ["A", "A", "B", "B", "C", "A", "C", "A", "B"],
    "Manager": ["-", "-", "-", "-", "-", "-", "-", "-", "-"],
    "Sales_bef": [132921, 149559, 146278, 152167, 159525, 137163, 130625, None, 131140],
    "Sales_After_digital_add": [270390, 223334, 244243, 231808, 258402, 256948, 222106, None, 226261]
}

df = pd.DataFrame(data)
df.dropna(inplace=True)
df['Increase_in_sales'] = df['Sales_After_digital_add'] - df['Sales_bef']

average_increase = df['Increase_in_sales'].mean()
print("Average increase in sales after digital marketing:", average_increase)
t_statistic, p_value = ttest_rel(df['Sales_After_digital_add'], df['Sales_bef'])
print("\nPaired t-test results:")
print("T-statistic:", t_statistic)
print("P-value:", p_value)


```

```python
crosstab = pd.crosstab(df['Region'], df['Manager'])

chi2, p, dof, expected = chi2_contingency(crosstab)

print("\nChi-square test results:")
print("Chi-square:", chi2)
print("P-value:", p)
```
